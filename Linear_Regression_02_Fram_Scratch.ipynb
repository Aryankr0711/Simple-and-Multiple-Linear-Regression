```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
class selfLR:
  def __init__(self):
    self.m=None
    self.b=None
    self.num=0
    self.den=0


  def fit(self, X_train, y_train):

    x_mean = X_train.mean()
    y_mean = y_train.mean()
    for i in range(X_train.shape[0]):
      self.num=self.num+((X_train[i]-x_mean) * (y_train[i]-y_mean))
      self.den=self.den+((X_train[i]-x_mean)**2)

    self.m= self.num/self.den
    self.b= y_train.mean()-(self.m*x_mean)


  def predict(self, X_test):
    return X_test*self.m+self.b
```

```python
lr=selfLR()
```

```python
df=pd.read_csv('train.csv')
```

```python
df.dropna(inplace=True)
```

```python
x=df['x']
y=df['y']
```

```python
X_train, X_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr.fit(X_train.values, y_train.values)
```

```python
lr.predict(8)
```
