```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df=pd.read_csv('Ice_cream selling data.csv')
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
sns.set_style(style='dark')
```

```python
sns.scatterplot(x=df['Temperature (°C)'], y=df['Ice Cream Sales (units)'])
plt.show()
```

```python
sns.regplot(x=df['Temperature (°C)'], y=df['Ice Cream Sales (units)'])
plt.show()
# data is not somewhat linear to show
```

```python
X_train, X_test, y_train, y_test=train_test_split(df['Temperature (°C)'],df['Ice Cream Sales (units)'],
                                                          test_size=0.2, random_state=42 )
```

```python
X_train.shape, X_test.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train.values.reshape(-1, 1), y_train.values)
```

```python
y_pred=lr.predict(X_test.values.reshape(-1, 1))
```

```python
mse= mean_squared_error(y_test, y_pred)
mae= mean_absolute_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2= r2_score(y_test, y_pred)


print("MSE: ", mse)
print("MAE: ", mae)
print("RMSE: ", rmse)
print("R^2: ", r2)
```

```python
sgd=SGDRegressor()
```

```python
sgd.fit(X_train.values.reshape(-1, 1), y_train.values)
```

```python
y_pred=sgd.predict(X_test.values.reshape(-1, 1))
```

```python
mse= mean_squared_error(y_test, y_pred)
mae= mean_absolute_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2= r2_score(y_test, y_pred)


print("MSE: ", mse)
print("MAE: ", mae)
print("RMSE: ", rmse)
print("R^2: ", r2)
```
