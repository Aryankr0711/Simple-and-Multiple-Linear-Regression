# 1. L2- Ridge Regularization

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df=load_diabetes()
```

```python
sns.kdeplot(df.data)
plt.show()
# Somewaht normal distribution
```

```python
sns.kdeplot(df.target)
plt.show()
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.data, df.target, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2= r2_score(y_test, y_pred)

print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
print('R2',r2)
```

```python
from sklearn.linear_model import Ridge
```

```python
lr=Ridge(alpha=0.001)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2= r2_score(y_test, y_pred)

print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
print('R2', r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.data, df.target, test_size=0.2, random_state=42)
```

```python
std=StandardScaler()
```

```python
X_train=std.fit_transform(X_train)
```

```python
X_test=std.transform(X_test)
```

```python
pr=PolynomialFeatures(degree=2)
```

```python
X_train=pr.fit_transform(X_train)
```

```python
X_test=pr.transform(X_test)
```

```python
lr=Ridge(alpha=0.001)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2= r2_score(y_test, y_pred)

print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
print('R2', r2)
```
